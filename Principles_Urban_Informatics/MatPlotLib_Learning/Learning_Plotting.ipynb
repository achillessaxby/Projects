```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
from collections import Counter
from itertools import chain

from mpl_toolkits.mplot3d import Axes3D
import numpy as np

%pylab inline
```

### Plot 1

```python
df = pd.read_csv('Mockaroo_Random_Data.csv')
df.head()
```

```python
df['Country'][df['Country'] == 'United States'] = None
df['Country'] = df['Country'].astype(float)
```

```python
df.head()
```

```python
df.fillna(value = 1, method = None, axis = 1, inplace = True)
df.head()
```

```python
df['Time-Zone'][:]
```

```python
df_counts = df['Time-Zone'].value_counts()
df_counts[:]
```

```python
ax = df_counts[:].plot(x='Time-Zone', y='Counts', kind='barh', figsize=(7, 7), 
                       fontsize=14, color='Red', rot = 0)
ax.set_xlabel("Total Counts of the Time-Zones", fontsize = 13, fontweight = 'bold')
ax.set_title("Plotting of the Total Time-Zones in America from Randomized Data"
             , fontsize = 13, fontweight = 'bold')
plt.text(0,-2.5, 'Figure 1 : The Total Time-Zones that have been collected from the Randomly Generated Data'
         ,fontsize = 10, fontweight = 'bold')
plt.legend(loc = 'upper right')
```

This is the plotting of different time-zones recovered from data that was randomly downlaoded and stored for further reference.

### Plot 2

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for c, z in zip(['r', 'g', 'b', 'y'], [30, 20, 10, 0]):
    xs = np.arange(20)
    ys = np.random.rand(20)

    # You can provide either a single color or an array. To demonstrate this,
    # the first bar of each set will be colored cyan.
    cs = [c] * len(xs)
    cs[0] = 'c'
    ax.bar(xs, ys, zs=z, zdir='y', color=cs, alpha=0.8)

ax.set_xlabel('X', fontsize = 11, fontweight = 'bold')
ax.set_ylabel('Y', fontsize = 11, fontweight = 'bold')
ax.set_zlabel('Z', fontsize = 11, fontweight = 'bold')
ax.set_title("Plotting Random Data", fontsize = 13, fontweight = 'bold')
plt.show()
```
