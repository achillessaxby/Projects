# Assignment - 1 - Solutions

```python
from __future__ import print_function
import pylab as pl
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit, minimize
%pylab inline

from scipy.stats import norm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

# Chi_Square Distribution

```python
mean_taken = 100
np.random.seed(5762)

chi_square = np.random.chisquare(mean_taken, 2000)
pl.hist(chi_square, bins = 26)
pl.xlabel('X')
pl.ylabel('N')
pl.title('Chi-Square', fontsize = 18, fontweight = 'bold')
pl.text(40,-50, 'Figure 1:', fontsize = 15, fontweight = 'bold')
pl.text(70,-50, 'Chi-Square Distribution : Histogram with mean at 100', fontsize = 14)

size = (1500 / np.array(range(1,101)))

chi_dist = []
for i in range(0, len(size)):
    chi_dist.append(np.random.chisquare(mean_taken, size[i]))

chi_means = []
for i in range(0 , len(size)):
    chi_means.append(chi_dist[i].mean())
    
chi_sq_plot = pl.figure(figsize = (10,6)).add_subplot(111)
chi_sq_plot.plot(size, chi_means, 'o')
chi_sq_plot.set_xlabel('Sample Size', fontsize = 18)
chi_sq_plot.set_ylabel('Sample Mean', fontsize = 18)
chi_sq_plot.set_title('Chi-Squared Distribution', fontsize = 18, fontweight = 'bold')
chi_sq_plot.text(0,89, 'Figure 2:', fontsize = 15, fontweight = 'bold')
chi_sq_plot.text(250,89, 'Chi-Square Distribution : Scatter Plot with Mean at 100', fontsize = 14)
chi_sq_plot.plot([min(size), max(size)], [mean_taken, mean_taken], 'k')
```

# Normal Distribution

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import math

mu = 100
stdev = np.std(size)

x = np.random.normal(mu, stdev, 2000)
count, bins, ignored = plt.hist(x, 26, normed=True)

plt.plot(10, 1/(stdev * np.sqrt(2 * np.pi)) * np.exp( - (10 - mu)**2 / (2 * stdev**2) ), linewidth=2)
plt.xlabel('X')
plt.ylabel('N')
pl.title('Normal', fontsize = 18, fontweight = 'bold')
pl.text(-800,-0.0004, 'Figure 3:', fontsize = 15, fontweight = 'bold')
pl.text(-250,-0.0004, 'Normal Distribution : Histogram with mean at 100', fontsize = 14)
plt.show()

normal_dist = []
for i in range(0, len(size)):
    normal_dist.append(np.random.normal(mean_taken, stdev, size[i]))

normal_means = []
for i in range(0 , len(size)):
    normal_means.append(normal_dist[i].mean())
```

```python
normal_dist_plot = pl.figure(figsize = (10,6)).add_subplot(111)
normal_dist_plot.plot(size, normal_means, 'o')
normal_dist_plot.set_xlabel('Sample Size', fontsize = 18)
normal_dist_plot.set_ylabel('Sample Mean', fontsize = 18)
normal_dist_plot.set_title('Normal Distribution', fontsize = 18 , fontweight = 'bold')
normal_dist_plot.text(0,-100, 'Figure 4:', fontsize = 15, fontweight = 'bold')
normal_dist_plot.text(300,-100, 'Normal Distribution : Scatter Plot with mean at 100', fontsize = 14)
normal_dist_plot.plot([min(size), max(size)], [mean_taken, mean_taken])
```

# Poisson Distribution

```python
lambda_poisson_mean = 100

s = np.random.poisson(lambda_poisson_mean, 2000)
count, bins, ignored = plt.hist(s, 26, normed=True)
plt.xlabel('X')
plt.ylabel('N')
pl.title('Poisson', fontsize = 18, fontweight = 'bold')
pl.text(70,-0.015, 'Figure 5:', fontsize = 15, fontweight = 'bold')
pl.text(90,-0.015, 'Poisson Distribution : Histogram with mean at 100', fontsize = 14)
plt.show()

poisson_dist = []
for i in range(0, len(size)):
    poisson_dist.append(np.random.poisson(lambda_poisson_mean, size[i]))

poisson_means = []
for i in range(0 , len(size)):
    poisson_means.append(poisson_dist[i].mean())
    
poisson_dist_plot = pl.figure(figsize = (10,6)).add_subplot(111)
poisson_dist_plot.plot(size, poisson_means, 'o')
poisson_dist_plot.set_xlabel('Sample Size', fontsize = 18)
poisson_dist_plot.set_ylabel('Sample Mean', fontsize = 18)
poisson_dist_plot.set_title('Poisson Distribution', fontsize = 18 , fontweight = 'bold')
poisson_dist_plot.text(0,92, 'Figure 6:', fontsize = 15, fontweight = 'bold')
poisson_dist_plot.text(300,92, 'Poisson Distribution : Scatter Plot with mean at 100', fontsize = 14)
poisson_dist_plot.plot([min(size), max(size)], [mean_taken, mean_taken])
```

# Binomial Distribution

```python
b = np.random.binomial(200, 0.5, 2000)

count, bins, ignored = plt.hist(b, 26, normed=True)
plt.xlabel('X')
plt.ylabel('N')
pl.title('Binomial', fontsize = 18, fontweight = 'bold')
pl.text(70,-0.015, 'Figure 7:', fontsize = 15, fontweight = 'bold')
pl.text(87,-0.015, 'Binomial Distribution : Histogram with mean at 100', fontsize = 14)
plt.show()

binomial_dist = []
for i in range(0, len(size)):
    binomial_dist.append(np.random.binomial(200, 0.5, 2000))

binomial_means = []
for i in range(0 , len(size)):
    binomial_means.append(binomial_dist[i].mean())
    
binomial_dist_plot = pl.figure(figsize = (10,6)).add_subplot(111)
binomial_dist_plot.plot(size, binomial_means, 'o')
binomial_dist_plot.set_xlabel('Sample Size', fontsize = 18)
binomial_dist_plot.set_ylabel('Sample Mean', fontsize = 18)
binomial_dist_plot.set_title('Binomial Distribution', fontsize = 18, fontweight = 'bold')
binomial_dist_plot.text(0,99.4,'Figure 8:', fontsize = 15, fontweight = 'bold')
binomial_dist_plot.text(300,99.4,'Binomial Distribution : Scatter Plot with mean at 100', fontsize = 14)
binomial_dist_plot.plot([min(size), max(size)], [mean_taken, mean_taken])
```

# Logistic Distribution

```python
l = np.random.logistic(mean_taken, 1.0, 2000)

count, bins, ignored = plt.hist(l, bins=50)
plt.xlabel('X')
plt.ylabel('N')
pl.title('Logistic', fontsize = 18, fontweight = 'bold')
pl.text(91,-45, 'Figure 9:', fontsize = 15, fontweight = 'bold')
pl.text(96,-45, 'Logistic Distribution : Histogram with mean at 100', fontsize = 14)
plt.show()

logistic_dist = []
for i in range(0, len(size)):
    logistic_dist.append(np.random.logistic(mean_taken, 1.0, 2000))

logistic_means = []
for i in range(0 , len(size)):
    logistic_means.append(logistic_dist[i].mean())
    
logistic_dist_plot = pl.figure(figsize = (10,6)).add_subplot(111)
logistic_dist_plot.plot(size, binomial_means, 'o')
logistic_dist_plot.set_xlabel('Sample Size', fontsize = 18)
logistic_dist_plot.set_ylabel('Sample Mean', fontsize = 18)
logistic_dist_plot.set_title('Logistic Distribution', fontsize = 18 , fontweight = 'bold')
logistic_dist_plot.text(0,99.45,'Figure 10:', fontsize = 15, fontweight = 'bold')
logistic_dist_plot.text(300,99.45,'Logistic Distribution : Scatter Plot with mean at 100', fontsize = 14)
logistic_dist_plot.plot([min(size), max(size)], [mean_taken, mean_taken])
```

```python
allmeans = []

allmeans.extend(chi_means)
allmeans.extend(normal_means)
allmeans.extend(binomial_means)
allmeans.extend(poisson_means)
allmeans.extend(logistic_means)

pl.figure(figsize=(10, 10))
pl.hist(allmeans, bins=100, normed=True)
pl.title('All means distribution', fontsize = 18, fontweight = 'bold')
pl.xlabel('Sample Mean', fontsize = 18)
pl.ylabel('Probablity', fontsize = 18)
pl.text(-50,-0.03, 'Figure 11:', fontsize = 15, fontweight = 'bold')
pl.text(-10,-0.03, 'Histogram - All Means Combined', fontsize = 14)
pl.show()
```
