```python
# The To-Do-List is of a different input kind - but it is a To-Do-List that solely depends on user interaction with the list.
# To Do !! We all have those days when there are a bajillion things to do, and we don't know how we're going to get it all done. 
# Sooner than later we miss the tasks that we need to do.
# Solution for Question-1(Non-Bonus Track):

def menuTask():
    print("1 : To Add a new entry to the list")
    print("2 : To Delete an entry from the list")
    print("3 : To View the list")
    print("4 : To Exit")

def viewTask():
    print("TO DO LIST\n")
        
    f = open('to_do_list.txt','r+')
    list = f.read()
    if list == '':
        print("No Tasks in the List\n")
        return
    f.close()
    lists = []
    lists = list.split('\n')
    print("The tasks in the To Do List are : \n")
    for index,lines in enumerate(lists):
        print("{} : {} \n".format(index+1,lines))
    
def addTask():
    entry = raw_input("Enter the item to the list: \n")
    f = open('to_do_list.txt','r+')
    list = f.read()
    lists = []
    lists = list.split('\n')
    if lists[0]=='':
        del lists[0]
    f.close()
    f = open('to_do_list.txt','w+')
    lists.append(entry)
    list = '\n'.join(lists)
    f.write(list)
    f.close()

def delTask():
    index = int(input("Enter the entry no which has to be deleted : \n"))
    f = open('to_do_list.txt','r+')
    list = f.read()
    lists = []
    lists = list.split('\n')
    f.close()
    try:
        del lists[index-1]
    except:
        print("Please enter the correct choice number\n")
    f =open('to_do_list.txt','w+')
    list = '\n'.join(lists)
    f.write(list)
    f.close()

def main():
    viewTask()
    menuTask()
    while True:
        try:
            choice = int(input("\nEnter your choice : \n"))
        except:
            print("The choice has to be an interger value\n")
            continue
        if choice==1:
            addTask()
        elif choice==2:
            delTask()
        elif choice==3:
            viewTask()
        else:
            exit()

main()
```

```python
# Solution for Question-2:

import math

mass = float(raw_input("Mass in Kilograms : "))
number = int(raw_input("Number of planets : "))

result = []

exp = (6.67e-11)

for i in range(number):
    input = raw_input("Enter Name, Radius(m) & Avg.Density(kg/m3) of planet: " ).split(',')    
    
    num = ((float(input[1])**3))*((float(4)/3)*math.pi)*(float(input[2]))*(float(mass))
    den = ((float(input[1])**2))

    F = float((num)*(exp))/(den)                                        

    result.append((input[0], F))

print("\nWeight in Newtons\n")    

for i in range(number):
    print "%s : %.3f" % (result[i][0],result[i][1])


'''
75
9
Mercury, 2439700, 5427
Venus, 6051900, 5243
Earth, 6367445, 5515
Mars, 3386000, 3934
Jupiter, 69173000, 1326
Saturn, 57316000, 687
Uranus, 25266000, 1270
Neptune, 24553000, 1638
Pluto, 1173000, 2050
'''
```
