```python
import os
import pylab as pl
import numpy as np
%pylab inline
pl.xkcd()
a=0.05
pl.plot(np.arange(20),1.-pow((1-a),np.arange(20)), label='probability of type I error')
pl.xlabel("number of p>0.05 experiments")
pl.ylabel("p(>)")
pl.legend()
pl.savefig(os.getenv('PUI15PLOTS')+"/dredging.png")

```
