```python
import os
import numpy as np

import pylabsetup
import pylab as pl
import json

s = json.load(open(os.getenv('PUI2016')+"/fbb_matplotlibrc.json"))

%pylab inline
s['font.family'] = ['sans-serif']
pylab.rcParams.update(s)

#pylab.rc['font.sans-serif'] = 'Helvetica'
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
#pylab.rc['text'] = {'usetex': True}


if os.getenv('PUI15PLOTS') is None:
    print ("Must set environmental variable PUI16PLOTS to point to a directory where to store plots")
```

```python
x=np.arange(-10,10,0.01)

gs = lambda x: np.exp((-(x)**2)/2)*1.0/np.sqrt(2*np.pi)

fig = pl.figure(4,figsize=(10,5))
fig.patch.set_facecolor('black')

ax = fig.add_subplot(111, axisbg='black')
ax.tick_params(color='gray', labelcolor='gray')
for spine in ax.spines.values():
        spine.set_edgecolor('gray')
ax.plot(x,gs(x),color='white',lw=3)
#ax.xaxis.grid(True, which="major", linestyle='-', color='white') 
#ax.yaxis.grid(True, which="major", linestyle='-', color='white') 

ax.fill_between(np.arange(-1,1,0.01),np.zeros(len(np.arange(-1,1,0.01))),gs(np.arange(-1,1,0.01)),color='IndianRed', 
                                alpha=0.5,lw=3,label=r'1$\sigma$ = 16\%')
ax.set_xlabel("effect",color='gray')
                
#pl.legend()
ax.set_ylabel("p",color='gray')
ax.set_ylim(-0.05,0.45)
ax.set_xlim(-9,9)
pl.text(4.5,0.31,r'1$\sigma$ = 68%', color='IndianRed', size=25)\

pl.savefig(os.getenv('PUI15PLOTS')+'/gaus1sig_1.png', facecolor=fig.get_facecolor(), edgecolor='none')

fig = pl.figure(5,figsize=(10,5))
fig.patch.set_facecolor('black')

ax = fig.add_subplot(111, axisbg='black')
ax.tick_params(color='gray', labelcolor='gray')
for spine in ax.spines.values():
        spine.set_edgecolor('gray')
ax.plot(x,gs(x),color='white',lw=3)
#ax.xaxis.grid(True, which="major", linestyle='-', color='white') 
#ax.yaxis.grid(True, which="major", linestyle='-', color='white') 

ax.fill_between(np.arange(-1,1,0.01),np.zeros(len(np.arange(-1,1,0.01))),gs(np.arange(-1,1,0.01)),color='IndianRed', 
                                alpha=0.5,lw=3,label=r'1$\sigma$ = 16%')
ax.fill_between(np.arange(-2,2,0.01),np.zeros(len(np.arange(-2,2,0.01))),gs(np.arange(-2,2,0.01)),color='IndianRed', 
                                alpha=0.5,lw=3,label=r'2$\sigma$ = 5%')
ax.set_xlabel("effect",color='gray')
                
#pl.legend()
ax.set_ylabel("p",color='gray')
ax.set_ylim(-0.05,0.45)
ax.set_xlim(-9,9)
pl.text(4.5,0.31,r'2$\sigma$ = 95%', color='IndianRed', size=25)
pl.savefig(os.getenv('PUI15PLOTS')+'/gaus1sig_2.png', facecolor=fig.get_facecolor(), edgecolor='none')


fig = pl.figure(6,figsize=(10,5))
fig.patch.set_facecolor('black')

ax = fig.add_subplot(111, axisbg='black')
ax.tick_params(color='gray', labelcolor='gray')
for spine in ax.spines.values():
        spine.set_edgecolor('gray')
ax.plot(x,gs(x),color='white',lw=3)
#ax.xaxis.grid(True, which="major", linestyle='-', color='white') 
#ax.yaxis.grid(True, which="major", linestyle='-', color='white') 

ax.fill_between(np.arange(-1,1,0.01),np.zeros(len(np.arange(-1,1,0.01))),gs(np.arange(-1,1,0.01)),color='IndianRed', 
                                alpha=0.5,lw=3)
ax.fill_between(np.arange(-2,2,0.01),np.zeros(len(np.arange(-2,2,0.01))),gs(np.arange(-2,2,0.01)),color='IndianRed', 
                                alpha=0.5,lw=3)
ax.fill_between(np.arange(-3,3,0.01),np.zeros(len(np.arange(-3,3,0.01))),gs(np.arange(-3,3,0.01)),color='IndianRed', 
                                alpha=0.5,lw=3)
ax.set_xlabel("effect",color='gray')
                
#pl.legend()
ax.set_ylabel("p",color='gray')
ax.set_ylim(-0.05,0.45)
ax.set_xlim(-9,9)
pl.text(4.5,0.31,r'3$\sigma$ = 99.7%', color='IndianRed', size=25)

pl.savefig(os.getenv('PUI15PLOTS')+'/gaus1sig_3.png', facecolor=fig.get_facecolor(), edgecolor='none')
```
