```python
import os
import pylab as pl
import numpy as np
import scipy 
%pylab inline 


import json
s = json.load( open("fbb_matplotlibrc.json") )
pl.rcParams.update(s)

%pylab inline

from numpy import prod
from scipy.stats import poisson




fig=pl.figure(figsize=(10,10))
ax = fig.add_subplot(111, axisbg='black')
for l in range(100):
    x=np.arange(150)
    #print scipy.misc.factorial(x)
    ax.plot(x, poisson.pmf(x, l), color='gray',alpha=0.5)

# set ticks and tick labels

# Only draw spine between the y-ticks
ax.spines['left'].set_bounds(0, 0.35)
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
# Only show ticks on the left and bottom spines
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
   

ax.spines['bottom'].set_color('gray')
#ax.spines['top'].set_color('red')
ax.xaxis.label.set_color('gray')
ax.tick_params(axis='y', colors='gray')
ax.tick_params(axis='x', colors='gray')
ax.spines['left'].set_color('gray')
#ax.spines['top'].set_color('red')
fig.patch.set_facecolor('black')

ax.yaxis.label.set_color('gray')
pl.grid(False)
pl.savefig(os.getenv('PUI15PLOTS')+"/poisson_bg.png", color='black',facecolor=fig.get_facecolor(), edgecolor='none')
    
```
