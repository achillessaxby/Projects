### demonstrates how to generate a Gaussian distribution with different moments

```python
import pylab as pl
import pandas as pd
import numpy as np
import os

%pylab inline


import json
s = json.load( open(os.getenv('PUI2016')+"/fbb_matplotlibrc.json") )
pl.rcParams.update(s)
```

```python
pl.hist(np.random.randn(1000))
```

# changing the mean

```python
pl.hist(np.random.randn(1000))
pl.hist(np.random.randn(1000) + 10)
```

# changing the standard deviation

```python
pl.hist(np.random.randn(1000) + 10)
pl.hist(np.random.randn(1000) * 10 + 10, alpha=0.5)
```

```python
pl.hist(np.random.randn(1000) + 10)
pl.hist(np.random.poisson(10, 1000), alpha=0.5)
```

# Other distributions: Cauchy

```python
pl.hist(np.random.randn(1000))
pl.hist(np.random.standard_cauchy(100), alpha = 0.5)
```
