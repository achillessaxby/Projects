```python
%matplotlib inline
import matplotlib.pyplot as plt

numbers = range(1,10)
print numbers

squares = map(lambda x: x**2, numbers)
print squares

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(numbers, squares)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.plot(numbers, squares)
```

```python
ax.set_title('My First Plot')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('My First Plot')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.plot(numbers, squares)
```

```python
cubes = map(lambda x: x**3, numbers)
print cubes
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('My First Plot')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.plot(numbers, squares)
ax.plot(numbers, cubes)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('My First Plot')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.plot(numbers, squares, label="Squares")
ax.plot(numbers, cubes, label="Cubes")
ax.legend(loc="Upper Left")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Squares and Cubes of Numbers')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.plot(numbers, squares, label="Squares", marker='o', linestyle='--', color='red')
ax.plot(numbers, cubes, label="Cubes", marker='+', linestyle=':', color='blue')
ax.legend(loc="upper left")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Squares and Cubes of Numbers')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.bar(numbers, squares, label="Squares", color='red')
ax.bar(numbers, cubes, label="Cubes", color='blue', alpha=0.2)
ax.legend(loc="upper left")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Squares and Cubes of Numbers')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.bar(numbers, squares, label="Squares", color='red')
ax.plot(numbers, squares)
ax.bar(numbers, cubes, label="Cubes", color='blue', alpha=0.2)
ax.plot(numbers, cubes)
ax.legend(loc="upper left")
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Squares and Cubes of Numbers')
ax.set_xlabel('Numbers')
ax.set_ylabel('Squares')
ax.bar(numbers, squares, label="Squares", color='red', align='center')
ax.plot(numbers, squares)
ax.bar(numbers, cubes, label="Cubes", color='blue', align='center',alpha=0.2)
ax.plot(numbers, cubes)
ax.legend(loc="upper left")
```
